Progressing to the selection of three random numbers from 1-10 with replacement

```python
#picks 3 random numbers
numbers = [
    random.randint(1, 10),
    random.randint(1, 10),
    random.randint(1, 10)
]

# player guesses 3 numbers
guess1 = int(input("Guess your first number (1‚Äì10): "))
guess2 = int(input("Guess your second number (1‚Äì10): "))
guess3 = int(input("Guess your third number (1‚Äì10): "))

guesses = [guess1, guess2, guess3]

# count how many are correct
correct = 0
for g in guesses:
    if g in numbers:
        correct += 1

# results
if correct == 3:
    print("üéâ Congratulations! All 3 numbers were correct!")
elif correct == 2:
    print("üòÑ Nice! You got 2 numbers correct out of",numbers)
elif correct == 1:
    print("üôÇ You got 1 number correct out of",numbers)
else:
    print("‚ùå Oops wrong. The numbers were:", numbers,"good luck next time")
```
